```python
import pandas as pd
```

```python
day1_df = pd.read_csv("data/day1.txt", names=["depths"])
day1_df
```

```python
day1_df["delta"] = day1_df.depths.diff()
day1_df
```

```python
sum(day1_df.delta > 0)
```

```python
day1_df["depths_smooth"] = day1_df.depths.rolling(window=3).sum()
day1_df["delta_smooth"] = day1_df.depths_smooth.diff()
```

```python
sum(day1_df.delta_smooth > 0 )
```
