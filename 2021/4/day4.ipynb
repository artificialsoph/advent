```python
curr_text = input_text
```

```python
draw_list = curr_text.split("\n\n")[0]
draw_list = [int(i) for i in draw_list.split(",")]
draw_arr = np.array(draw_list)
```

```python
board_list = curr_text.split("\n\n")[1:]

board_list = [
    [[int(i) for i in row.split()] for row in board.split("\n")] for board in board_list
]
board_arr = np.array(board_list)
```

```python
marked_arr = np.zeros_like(board_arr, dtype=bool)

for num in draw_arr:

    print(f"drawing {num}")

    marked_arr[board_arr == num] = True

    board_win = marked_arr.all(axis=2).any(axis=1) + marked_arr.all(axis=1).any(axis=1)

    if board_win.any():
        board_i = np.argwhere(board_win).flat[0]

        unmarked_nums = board_arr * np.logical_not(marked_arr)

        winning_board = unmarked_nums[board_i, ...]

        score = np.sum(winning_board)

        final_score = num * score
        print(final_score)

        break
```

# Part Two

```python
marked_arr = np.zeros_like(board_arr, dtype=bool)

board_state_hist = np.zeros([draw_arr.shape[0], board_arr.shape[0]])

for i, num in enumerate(draw_arr):

    print(f"drawing {num}")

    marked_arr[board_arr == num] = True

    board_state_hist[i, :] = marked_arr.all(axis=2).any(axis=1) + marked_arr.all(
        axis=1
    ).any(axis=1)

    num_winning_boards = sum(board_state_hist[i, :])

    print(f"{num_winning_boards} boards winning")

    if board_state_hist[i, :].all():
        # Warning, assumes there is a single, last board to win.

        last_board_i = np.argwhere(board_state_hist[i - 1, :] == 0).flat[0]

        unmarked_nums = board_arr * np.logical_not(marked_arr)

        winning_board = unmarked_nums[last_board_i, ...]
        print(winning_board)

        score = np.sum(winning_board)
        print(score)

        final_score = num * score
        print(final_score)

        break
```
